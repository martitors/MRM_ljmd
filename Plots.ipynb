```python
import numpy as np
import matplotlib.pyplot as plt
from cycler import cycler

from matplotlib.ticker import ScalarFormatter
```

```python
############################## SERIAL PLOT ##################################

n_opts = [1,2,4,8,16]
names = ['Baseline','\n-O3','More Optimization Flag','\nNewtons 3rd law','No expensive math']
#fake times, you need to replace this with values
times_108 = [16.303,12.587,2.958,1.29,1.46] 
times_2916 = [300.307,169.68,100.272,48.614,46.378]
times_78732 = [2223.907,766.217,743.296,389.974,370.301]

bar_width = 0.2
bar_positions1 = np.arange(len(n_opts))
bar_positions2 = bar_positions1 + bar_width
bar_positions3 = bar_positions2 + bar_width

plt.rcParams['axes.prop_cycle'] = cycler('color', plt.get_cmap('Paired').colors[::2])

plt.figure(figsize=(10,5))
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.bar(bar_positions1,[times_108[0]/t for t in times_108],label = "108 atoms", width=bar_width,edgecolor = "black")
plt.bar(bar_positions2,[times_2916[0]/t for t in times_2916],label = "2916 atoms", width=bar_width,edgecolor = "black")
plt.bar(bar_positions3,[times_78732[0]/t for t in times_78732],label = "78732 atoms", width=bar_width,edgecolor = "black")
plt.xticks(bar_positions2,names)
plt.title('SERIAL Benchmark',fontsize = 12)
plt.ylabel("Speedup (log)")
plt.yscale('log')
plt.legend()
plt.savefig('plots/Serial_bm_1.png',format='png')


plt.figure(figsize=(10,5))
plt.plot(bar_positions1,times_108,'o-',label = "108 atoms",linewidth=2.5)
plt.plot(bar_positions1,times_2916,'o-',label = "2916 atoms",linewidth=2.5)
plt.plot(bar_positions1,times_78732,'o-',label = "78732 atoms",linewidth=2.5)
plt.xticks(bar_positions1,names)
plt.title('SERIAL Benchmark',fontsize = 12)
plt.ylabel("Time (log s)")
plt.yscale('log')
plt.legend()
plt.savefig('plots/Serial_bm_2.png',format='png')


plt.show()
```
